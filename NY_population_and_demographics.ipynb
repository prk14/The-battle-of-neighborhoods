```python
from bs4 import BeautifulSoup
import csv
print('All required libraries are imported')
```

```python
url=requests.get('https://en.wikipedia.org/wiki/Demographics_of_New_York_City').text
soup=BeautifulSoup(url,'lxml')
table = soup.find('table',{'class':'wikitable sortable'})
headers = [header.text for header in table.find_all('th')]
table_rows = table.find_all('tr')
rows = []
for row in table_rows:
   td = row.find_all('td')
   row = [row.text for row in td]
   rows.append(row)

# write these rows and headers in a csv file
with open('Ny_population.csv','w') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(row for row in rows if row)
```

```python
# load the data from csv to dataframe
df=pd.read_csv('Ny_population.csv')
df.head()
```

```python
# first remove columns after county as they are NaN
df.drop(['Estimate (2019)','billions(2012 US$)','square miles','squarekm','persons /mi2','persons /km2\r\n'],axis=1, inplace=True)
df.head()
```

```python
#get column names properly
df.columns = df.columns.str.replace(' ','')
df.columns = df.columns.str.replace('\'','')
df.head()
```

```python
df.rename(columns = {'NewYorkCitysfiveboroughsvte\r\n' : 'NY_Borough',
                   'Jurisdiction\r\n':'Jurisdiction',
                   'Population\r\n':'Population',
                    'GDP\r\n':'GDP',
                   'Landarea\r\n':'square_miles',
                    'Density\r\n':'square_km'}, inplace=True)
df
```

### Data cleaning

```python
df['NY_Borough']=df['NY_Borough'].replace(to_replace='\r', value='', regex=True)
df['Jurisdiction']=df['Jurisdiction'].replace(to_replace='\r', value='', regex=True)
df['Population']=df['Population'].replace(to_replace='\r', value='', regex=True)
df['GDP']=df['GDP'].replace(to_replace='\r', value='', regex=True)
df['square_miles']=df['square_miles'].replace(to_replace='\r', value='', regex=True)
df['square_km']=df['square_km'].replace(to_replace='\r', value='', regex=True)
df['Borough']=df['Borough'].replace(to_replace='\r', value='', regex=True)
df['County']=df['County'].replace(to_replace='\r', value='', regex=True)
df
```

```python
df['NY_Borough']=df['NY_Borough'].replace(to_replace='\n', value='', regex=True)

df
```

```python
df.fillna('')
df.head()
```

```python
# save df to csv
df.to_csv('ny_population_clean.csv')
```

### Demographic data

```python
# we will get demographic data from https://en.wikipedia.org/wiki/New_York_City
url = requests.get('https://en.wikipedia.org/wiki/New_York_City').text
soup = BeautifulSoup(url,'lxml')
table = soup.find_all('table',{'class':'wikitable collapsible collapsed'})
headers = [header.text for header in table[1].find_all('th')]

table_rows = table[1].find_all('tr')        
rows = []
for row in table_rows:
   td = row.find_all('td')
   row = [row.text for row in td]
   rows.append(row)

with open('NY_demographic.csv', 'w') as f:
   writer = csv.writer(f)
   writer.writerow(headers)
   writer.writerows(row for row in rows if row)
```

```python
demo_df=pd.read_csv('NY_demographic.csv')
demo_df.head()
```
